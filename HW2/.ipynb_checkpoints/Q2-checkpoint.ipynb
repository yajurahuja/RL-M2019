```python
import numpy as np
```

```python
V = np.zeros(25) # Value Matrix (25 X 1)
a = 1 ; a_ = 21 
b = 3 ; b_ = 13
delta = 0.9
```

```python
#Probability Matrix Population
P = np.zeros((25, 25)) #Probability Matrix (P(s'|s)) (25 X 25)
for i in range(25):
        if i == a: #Special State
            P[i][a_] = 1
        elif i == b: #Special State
            P[i][b_] = 1
        else:
            u_x = i//5
            u_y = i%5
            if u_x - 1 >= 0:
                P[i][(u_x - 1)* 5 + u_y] = 0.25
            else: 
                P[i][i] += 0.25
                
            if u_y - 1 >= 0:
                P[i][(u_x )* 5 + u_y - 1] = 0.25
            else: 
                P[i][i] += 0.25
                
            if u_x + 1 <= 4:
                P[i][(u_x + 1)* 5 + u_y] = 0.25 
            else: 
                P[i][i] += 0.25
                
            if u_y + 1 <= 4:
                P[i][(u_x)* 5 + u_y + 1] = 0.25
            else: 
                P[i][i] += 0.25
            
            
            
                
print(P)
print(np.sum(P, axis = 1))
        
        
```

```python
R = np.zeros(25) # Expected Reward Matrix (25 X 1)
print(R.reshape(5,5))
for i in range(25):
    if i == a:
        R[i] = 10
    elif i == b:
        R[i] = 5
    else:
        u_x = i//5
        u_y = i%5
       
        if u_x - 1 < 0:
            R[i] = R[i]-(1 * 0.25)
        if u_y - 1 < 0:
            R[i] = R[i]-(1 * 0.25)
        if u_x + 1 > 4:
            R[i] = R[i]-(1 * 0.25)
        if u_y + 1 > 4:
            R[i] = R[i]-(1 * 0.25)
        #print(R[i])   

print(R.reshape(5,5))
    
```

```python
V = np.matmul(np.linalg.inv(np.identity(25) - (delta*P)), R)
np.around(V.reshape(5,5), decimals = 1)
```
